```python
#import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potental energy
def potential_energy(x,k=1.0):
    return 0.5 * k * x**2 # k is the spring constant

#test function for the potential energy
potential_energy(2,10)
```

```python
#define mash for plotting 
x=np.linspace(-5,5,100) # np.linspace(x_min, x_max, npts)
y=potential_energy(x)

#testing the output
#print(np.transpose([x,y]))
```

```python
# plot the function
plt.figure(figsize=(5,4))  # Close the parenthesis here
plt.plot(x, y, color="black", linewidth=0.5)

# add axes labels
plt.xlabel("displacement (arbitrary unit)")
plt.ylabel("potential energy (arbitrary unit)")

# show the plot
plt.show()  # Combine all elements of the plot
```

```python
#import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potental energy
def potential_energy_2D(x,y,kx=1.0,ky=1.0):
    return 0.5 * (kx * x**2 + ky * y**2) # k is the spring constant

#test function for the potential energy
potential_energy_2D(2,0)
```

```python
#define mash for plotting 
x=np.linspace(-5,5,3) # np.linspace(x_min, x_max, npts)
y=np.linspace(-5,5,3) # np.linspace(y_min, y_max, npts)

#combine 1D meshs into meshgrid
X, Y = np.meshgrid(x,y)

# define Z values
Z = potential_energy_2D(X,Y)

# testing the mesh
print(X)
print()
print(Y)
print()
print(Z)
```

```python
# construct the contour plot
plt.figure(figsize=(4,3))  # Correct the figsize syntax
plt.contour(X, Y, Z, levels=50)

# add a colorbar legend
plt.colorbar(label='potential energy (a)')

# add axes labels
plt.xlabel("x-displacement (arbitrary unit)")
plt.ylabel("y-displacement (arbitrary unit)")

# change limits of x and y
plt.xlim([-4,4])
plt.ylim([-4,4])

plt.show()  # Combine all plot objects
```
